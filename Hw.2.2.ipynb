```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
""" Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum) """
print(table.isnull().sum())
```

```python
"""Видаліть усі стовпці з пропусками, крім стовпця "Мова програмування"
Знову порахуйте, яка частка пропусків міститься в кожній колонці і переконайтеся, що залишився тільки стовпець "Мова.програмування"
Видаліть усі рядки у вихідній таблиці за допомогою методу dropna
Визначте новий розмір таблиці за допомогою методу shape"""

print("\n<-------------------------------------------------------------------------------------------->\n")
table = table.drop(["Спеціалізація", "Університет", "Предметна область", "Валюта", "cls"], axis=1)
print(table.isnull().sum())
print("\n<-------------------------------------------------------------------------------------------->\n")
table = table.dropna()
print(table.shape)
print("\n<-------------------------------------------------------------------------------------------->\n")
print(table.isnull().sum())
```

```python
""" Створіть нову таблицю python_data, в якій будуть тільки рядки зі спеціалістами, які вказали мову програмування Python
Визначте розмір таблиці python_data за допомогою методу shape """
python_data = table[table["Мова програмування"] == "Python"]
print(python_data.shape)
print("\n<-------------------------------------------------------------------------------------------->\n")
print(python_data)
```

```python
""" Використовуючи метод groupby, виконайте групування за стовпчиком "Посада" """
print("\n<-------------------------------------------------------------------------------------------->\n")
python_data.groupby(["Посада"])
print(python_data)
print("\n<-------------------------------------------------------------------------------------------->\n")
```

```python
python_data = python_data.agg({"Зарплата на місяць": ['min', 'max']})
print(python_data)
```

```python
def fill_avg_salary(row):
    return row["Зарплата на місяць"].mean()
python_data['avg'] = python_data.apply(fill_avg_salary, axis=1)
print(python_data['avg'].describe(include='all'))
```

```python
python_data.to_csv('python_data.csv', index=False)
```

```python
""" Гістограма розподілу зарплати """

table_python = table[table["Мова програмування"] == "Python"]
plt.figure(figsize=(14, 4))
sns.histplot(table_python['Зарплата на місяць'], bins=20, kde=True, color='skyblue')
plt.title('Розподіл зарплати серед спеціалістів Python')
plt.xlabel('Зарплата на місяць')
plt.ylabel('Частота')
plt.show()
```

```python
""" Діаграма розсіювання між досвідом роботи та зарплатою """

plt.figure(figsize=(14, 6))
sns.scatterplot(x='Досвід роботи на поточному місці', y='Зарплата на місяць', data=table_python, hue='Тип компанії', palette='viridis')
plt.title("Зв'язок між досвідом роботи та зарплатою")
plt.xlabel('Досвід роботи на поточному місці')
plt.ylabel('Зарплата на місяць')
plt.legend(title='Тип компанії')
plt.show()
```

```python
plt.figure(figsize=(14, 6))
sns.lineplot(x='Досвід роботи на поточному місці', y='Зарплата на місяць', data=table_python, marker='o', color='green')
plt.title('Залежність зарплати від досвіду роботи')
plt.xlabel('Досвід роботи на поточному місці')
plt.ylabel('Зарплата на місяць')
plt.show()
```
